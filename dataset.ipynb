Датасет можно скачать с по ссылке https://github.com/yutkin/Lenta.Ru-News-Dataset/releases/download/v1.1/lenta-ru-news.csv.bz2

```python
import bz2

with open("lenta-ru-news.csv.bz2", mode="+rb") as file:
    decompressed = bz2.decompress(file.read())
```

```python

with open("lenta-ru-news.csv", mode="wb") as new_file:
    new_file.write(decompressed)
```

```python
del(decompressed)
del(file)
del(new_file)
```

```python
import pandas as pd

full_ds = pd.read_csv("lenta-ru-news.csv")
```

```python
full_ds.head()
```

```python
filterred_ds = full_ds[full_ds["title"].str.contains(r"\d{3,}") == True]
filterred_ds.head(100)
```

```python
shaked_ds = filterred_ds.sample(n=1000)
shaked_ds.to_csv("news_with_numbers.csv")
```
